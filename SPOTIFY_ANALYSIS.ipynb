# Data Cleaning

**Create a copy of base data for manupulation & processing**

```python
spotify_data = spotify_analysis.copy()
```

```python
spotify_data.info()
```

```python
spotify_data
```

```python
spotify_data.drop(columns=['id'], axis = 1, inplace=True)
spotify_data.head()
```

**Extracting year from release date**

```python
spotify_data['year'] = pd.DatetimeIndex(spotify_data['release_date']).year
spotify_data
```

**Convert the Duration of the Songs From Milliseconds to Seconds.**

```python
spotify_data['duration'] = spotify_data['duration_ms'].apply (lambda x : round(x/1000))
spotify_data.drop('duration_ms', inplace = True, axis=1)
spotify_data.duration.head()
```

**Correlation Map**

```python
sd = spotify_data.drop(['name','artists','id_artists','release_date'], axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))
map = sns.heatmap(sd, annot = True, cmap='coolwarm')
map.set_title('Correlation Heatmap between Variable')
```

After running the code, we got our correlation map. On the right side, you can see a scale ranging from -0.6 to +1. Here -0.6 denotes the variables that have the least or negative correlation, while the values above 0.0 denote the variables with a positive correlation.

**Sample Only 4 Percent of the Whole Dataset**

```python
sample = spotify_data.sample(int(0.004 * len(spotify_data)))
print(len(sample))
```

This sample provides us 4 percent of the whole dataset that is 2346 rows. 

**Create a Regression Plot Between Loudness and Energy**

```python
plt.figure(figsize=(10,6))
reg = sns.regplot(data=sample, y='loudness', x='energy', color='Red')
reg.set_title('Loudness vs Energy Correlation')
```

1. There is a very high positive correlation between loudness and energy.
2. You can also see that all the data points or the songs are in one direction.
3. If the energy increases, the loudness of the song increases and similarly, if the song's loudness decreases, the energy of the track       also decreases.

**Create a Regression Plot Between Popularity and Acousticness**

```python
plt.figure(figsize=(10,6))
reg = sns.regplot(data = sample, y = "popularity", x = "acousticness", color = "b")
reg.set(title = "Popularity vs Acousticness Correlation")
```

we can see the blue color regression line is in downward direction, which denotes if the acousticness of the song increases, the popularity decreases and similarly, if the popularity increases, the acousticness decreases.
